Create a histogram for the following: hp,mpg,time-to-60,weight,cubicinches,year with respect to brand.Make sure to plot all the plots in the same plane with x and y label evenly spaced.

```python
fig = plt.subplots(2,3,figsize=(20,10))

brands = df['brand'].unique()

plt.subplot(2,3,1)    #plot-1
for brand in brands:
    plt.hist(x=df[df['brand'] == brand]['hp'], alpha=0.5, label=brand, edgecolor = 'black')
plt.xlabel('HP')
plt.title('Histogram of HP')
plt.legend()

plt.subplot(2,3,2)    #plot-2
for brand in brands:
    plt.hist(x=df[df['brand'] == brand]['mpg'], alpha=0.5, label=brand, edgecolor = 'black')
plt.xlabel('MPG')
plt.title('Histogram of MPG')
plt.legend()

plt.subplot(2,3,3)    #plot-3
for brand in brands:
    plt.hist(x=df[df['brand'] == brand]['time-to-60'], alpha=0.5, label=brand, edgecolor = 'black')
plt.xlabel('Time-to-60')
plt.title('Histogram of Time-to-60')
plt.legend()

plt.subplot(2,3,4)    #plot-4
for brand in brands:
    plt.hist(x=df[df['brand'] == brand]['weightlbs'], alpha=0.5, label=brand, edgecolor = 'black')
plt.xlabel('Weight')
plt.title('Histogram of Weight')
plt.legend()

plt.subplot(2,3,5)    #plot-5
for brand in brands:
    plt.hist(x=df[df['brand'] == brand]['cubicinches'], alpha=0.5, label=brand, edgecolor = 'black')
plt.xlabel('Cubic Inches')
plt.title('Histogram of Cubic Inches')
plt.legend()


plt.subplot(2,3,6)    #plot-6
for brand in brands:
    plt.hist(x=df[df['brand'] == brand]['year'], alpha=0.5, label=brand, edgecolor = 'black')
plt.xlabel('Year')
plt.title('Histogram of Year')
plt.legend()


plt.tight_layout()
plt.show()
```
